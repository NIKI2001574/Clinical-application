```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Choose top 20 keywords from previous analysis
top_20_keywords = [kw for kw, _ in keyword_freq.most_common(20)]

# Initialize a dataframe to hold binary flags for keyword presence per row
for kw in top_20_keywords:
    df[kw] = df['keywords'].fillna('').str.lower().apply(lambda x: 1 if kw in x else 0)

# Create a pivot table: rows are specialties, columns are keywords, values are mean occurrence rate
pivot = df.groupby('medical_specialty')[top_20_keywords].mean()

# Correlation heatmap
plt.figure(figsize=(16,10))
sns.heatmap(pivot, cmap='coolwarm', annot=True, fmt=".2f", linewidths=0.5)
plt.title('Mean Keyword Presence by Medical Specialty')
plt.xlabel('Keywords')
plt.ylabel('Medical Specialty')
plt.show()
```

```python
from spacy.lang.en.stop_words import STOP_WORDS

# Clean text
nlp = spacy.load("en_core_web_sm")

def clean_text(text):
    text = re.sub(r'\d+', '', str(text))
    text = re.sub(r'\s+', ' ', text).strip()
    doc = nlp(text)
    return ' '.join(
        token.lemma_.lower()
        for token in doc
        if token.text.lower() not in STOP_WORDS and token.is_alpha
    )

df['cleaned_transcription'] = df['transcription'].apply(clean_text)

# Save cleaned dataset
df_cleaned = df[['cleaned_transcription', 'medical_specialty']]
df_cleaned.to_csv("cleaned_medical_notes.csv", index=False)
print(" Cleaned dataset saved as 'cleaned_medical_notes.csv'")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import numpy as np

# Encode labels
label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['medical_specialty'])

# TF-IDF Vectorization (keep high dimensionality, no PCA)
tfidf_vectorizer = TfidfVectorizer(
    max_features=20000,       # Increased features
    ngram_range=(1, 3),       # Include unigrams, bigrams, and trigrams
    min_df=2,                 # Ignore very rare terms
    sublinear_tf=True         # Apply sublinear term frequency scaling
)
tfidf_matrix = tfidf_vectorizer.fit_transform(df['cleaned_transcription'])

# Train/test split with stratification
X_train, X_test, y_train, y_test = train_test_split(
    tfidf_matrix,
    df['label'],
    test_size=0.2,
    random_state=42,
    stratify=df['label']
)

# Logistic Regression with hyperparameter tuning
param_grid = {
    'C': [0.1, 1, 10],               # Regularization strength
    'solver': ['liblinear', 'saga'], # Solvers suitable for sparse data
    'class_weight': [None, 'balanced']
}

grid = GridSearchCV(
    LogisticRegression(max_iter=2000, random_state=42),
    param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)
grid.fit(X_train, y_train)

# Best model
best_lr = grid.best_estimator_
print("Best Parameters:", grid.best_params_)

# Predict
y_pred = best_lr.predict(X_test)

# Classification report
print(classification_report(y_test, y_pred, target_names=label_encoder.classes_))

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred, labels=range(len(label_encoder.classes_)))
plt.figure(figsize=(14, 10))
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=label_encoder.classes_)
disp.plot(cmap='Blues', xticks_rotation=45, ax=plt.gca())
plt.title("Confusion Matrix - Optimized Logistic Regression")
plt.tight_layout()
plt.show()
```

```python
import os
os.environ["USE_TF"] = "0"

#  Imports
import torch
from torch.utils.data import Dataset
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification, Trainer, TrainingArguments
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import pandas as pd

# Load and prepare dataset
df = pd.read_csv("cleaned_medical_notes.csv") 

# Filter top 5 specialties
top_categories = df['medical_specialty'].value_counts().head(5).index
df = df[df['medical_specialty'].isin(top_categories)].dropna(subset=['cleaned_transcription', 'medical_specialty'])

#  Encode labels
label_encoder = LabelEncoder()
df['label'] = label_encoder.fit_transform(df['medical_specialty'])

#  Train/test split
train_texts, test_texts, train_labels, test_labels = train_test_split(
    df['cleaned_transcription'].tolist(),
    df['label'].tolist(),
    test_size=0.2,
    random_state=42
)

#  Tokenizer (DistilBERT)
tokenizer = DistilBertTokenizer.from_pretrained('distilbert-base-uncased')
train_encodings = tokenizer(train_texts, truncation=True, padding=True, max_length=256)
test_encodings = tokenizer(test_texts, truncation=True, padding=True, max_length=256)

#  PyTorch Dataset
class MedNotesDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels
    
    def __getitem__(self, idx):
        return {
            'input_ids': torch.tensor(self.encodings['input_ids'][idx]),
            'attention_mask': torch.tensor(self.encodings['attention_mask'][idx]),
            'labels': torch.tensor(self.labels[idx])
        }

    def __len__(self):
        return len(self.labels)

train_dataset = MedNotesDataset(train_encodings, train_labels)
test_dataset = MedNotesDataset(test_encodings, test_labels)

#  Model (DistilBERT)
model = DistilBertForSequenceClassification.from_pretrained(
    'distilbert-base-uncased',
    num_labels=len(label_encoder.classes_)
)

#  Training arguments
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=1,  # Faster
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    warmup_steps=0,
    weight_decay=0.01,
    logging_dir='./logs',
    logging_steps=10,
    save_strategy='no'
)

#  Metrics
def compute_metrics(p):
    preds = torch.argmax(torch.tensor(p.predictions), axis=1)
    acc = accuracy_score(p.label_ids, preds)
    return {"accuracy": acc}

#  Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    compute_metrics=compute_metrics
)

#  Train
trainer.train()

#  Evaluate
eval_results = trainer.evaluate()
print(" Evaluation Results:", eval_results)

#  Predict + Classification Report
preds_output = trainer.predict(test_dataset)
preds = torch.argmax(torch.tensor(preds_output.predictions), axis=1)

print("\n Classification Report:")
print(classification_report(test_labels, preds, target_names=label_encoder.classes_))
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np
import torch

# Extract embeddings from the trained model
model.eval()
with torch.no_grad():
    outputs = model(
        input_ids=torch.tensor(test_encodings['input_ids']),
        attention_mask=torch.tensor(test_encodings['attention_mask'])
    )
    embeddings = outputs.logits.cpu().numpy()  # Ensure it's on CPU

# Apply t-SNE for dimensionality reduction to 2D
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
embeddings_2d = tsne.fit_transform(embeddings)

# Ensure test_labels is a flat integer array
test_labels = np.array(test_labels).astype(int).ravel()

# Plot the t-SNE visualization
plt.figure(figsize=(10, 7))
scatter = plt.scatter(
    embeddings_2d[:, 0], embeddings_2d[:, 1],
    c=test_labels, cmap='tab10', alpha=0.7
)

# Convert classes to a list to avoid ValueError
legend_labels = list(label_encoder.classes_)

plt.legend(
    handles=scatter.legend_elements()[0],
    labels=legend_labels,
    title="Specialty",
    loc="best"
)

plt.title("t-SNE Visualization of DistilBERT Test Embeddings")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Prepare data manually extracted from your reports for common specialties

# Logistic Regression metrics (subset of classes for clarity)
lr_metrics = {
    'Cardiovascular / Pulmonary': {'precision': 0.44, 'recall': 0.30, 'f1': 0.35},
    'Consult - History and Phy.': {'precision': 0.32, 'recall': 0.33, 'f1': 0.33},
    'Orthopedic': {'precision': 0.31, 'recall': 0.06, 'f1': 0.10},
    'Radiology': {'precision': 0.28, 'recall': 0.24, 'f1': 0.26},
    'Surgery': {'precision': 0.44, 'recall': 0.48, 'f1': 0.46},
}

# DistilBERT metrics
distilbert_metrics = {
    'Cardiovascular / Pulmonary': {'precision': 0.38, 'recall': 0.04, 'f1': 0.08},
    'Consult - History and Phy.': {'precision': 0.79, 'recall': 0.90, 'f1': 0.84},
    'Orthopedic': {'precision': 0.00, 'recall': 0.00, 'f1': 0.00},
    'Radiology': {'precision': 0.55, 'recall': 0.76, 'f1': 0.64},
    'Surgery': {'precision': 0.71, 'recall': 0.99, 'f1': 0.82},
}

# Convert to DataFrame for plotting
specialties = list(lr_metrics.keys())
metrics = ['precision', 'recall', 'f1']

data = []
for metric in metrics:
    lr_vals = [lr_metrics[s][metric] for s in specialties]
    distil_vals = [distilbert_metrics[s][metric] for s in specialties]
    for i, spec in enumerate(specialties):
        data.append({'Specialty': spec, 'Metric': metric.capitalize(), 'Model': 'Logistic Regression', 'Score': lr_vals[i]})
        data.append({'Specialty': spec, 'Metric': metric.capitalize(), 'Model': 'DistilBERT', 'Score': distil_vals[i]})

df_plot = pd.DataFrame(data)

# Plot grouped bar charts for each metric
fig, axes = plt.subplots(1, 3, figsize=(18,6), sharey=True)

for i, metric in enumerate(metrics):
    ax = axes[i]
    subset = df_plot[df_plot['Metric'] == metric.capitalize()]
    width = 0.35  # width of bars
    x = np.arange(len(specialties))
    
    lr_scores = subset[subset['Model']=='Logistic Regression']['Score'].values
    distil_scores = subset[subset['Model']=='DistilBERT']['Score'].values
    
    ax.bar(x - width/2, lr_scores, width, label='Logistic Regression', color='skyblue')
    ax.bar(x + width/2, distil_scores, width, label='DistilBERT', color='salmon')
    
    ax.set_xticks(x)
    ax.set_xticklabels(specialties, rotation=45, ha='right')
    ax.set_ylim(0,1)
    ax.set_title(metric.capitalize())
    ax.set_ylabel('Score')
    ax.grid(axis='y', linestyle='--', alpha=0.7)
    if i == 2:
        ax.legend(loc='upper left')

plt.suptitle('Model Performance Comparison by Medical Specialty', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
